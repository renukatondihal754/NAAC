# basic analysis of NAAC dataset

```python
import pandas as pd

#load the dataset
df = pd.read_csv("naac.csv", skiprows=4)
df.head(10)
```

```python
df.info()
```

```python
df.columns
```

```python
df = df[['State','Name of the College','CGPA','Grade','Accreditation valid up to']]
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
df['State'].value_counts().plot(kind="bar")
```

```python
df.sort_values(by=['CGPA'], ascending=False)
```

```python
df = df.join(df['Name of the College'].str.split(',', 1, expand=True).rename(columns={0:'Name', 1:'Address'}))
```

```python
df.drop("Name of the College",axis=1,inplace=True)
```

```python
df
```

```python
df[df['CGPA']>2.5]
```

```python
df[(df['CGPA']>2.5) & (df['Grade'] == 'B')]
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df['Grade'])
```

```python
df['New_Grade'] = le.transform(df['Grade'])
```
